#**DATA VISULLIZATION USING SEABORN**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
full_data=pd.read_csv('/content/titanic_dataset.csv')
```

```python
full_data.head()
```

```python
full_data.info()
```

##**Line Plot**

```python
sns.lineplot(data= full_data, x='Age', y='PassengerId')
```

```python
# Grouping data with ‘hue’
sns.lineplot(data= full_data, x='Age', y='PassengerId', hue='Survived')
```

```python
# Adding Styles
sns.lineplot(data= full_data, x='Age', y='PassengerId', hue='Sex', style='Sex')
```

```python
# Adding Markers
sns.lineplot(data= full_data, x='Age', y='PassengerId', hue='Sex', style='Sex', markers=True)
```

##**Bar Plot**

```python
sns.barplot(data= full_data, x='Sex', y='Survived')
```

```python
sns.barplot(data= full_data, x='Pclass', y='Survived')
```

```python
sns.barplot(data= full_data, x='Pclass', y='Survived', hue='Sex')
```

```python
custom_palette=['#2B1B17', 'red']
sns.barplot(data= full_data, x='Pclass', y='Survived', hue='Sex', palette=custom_palette, color='orange')
```

```python
sns.barplot(data= full_data, x='Pclass', y='Survived', hue='Sex', color='orange')
```

##**Scatterplot**

```python
sns.scatterplot(data=full_data, x='Fare', y='Age')
```

```python
#Here are some of the built-in palette options available in Seaborn:
#Qualitative Palettes:             'deep', 'pastel', 'dark', 'muted', 'bright', 'colorblind', etc.
#Sequential Palettes:              'viridis', 'plasma', 'inferno', 'magma', 'cividis', 'rocket', 'mako', 'icefire', etc.
#Diverging Palettes:               'vlag', 'twilight', 'twilight_shifted', 'hsv', 'seismic', etc.
sns.scatterplot(data=full_data, x='Fare', y='Age', hue='Pclass', style='Pclass', palette='viridis')
```

##**Histogram/Distplot**

```python
sns.distplot(full_data['Parch'])
```

```python
sns.distplot(full_data['Parch'], hist=False)
```

```python
sns.distplot(full_data['Parch'], hist=False, color='r')
```

```python
sns.distplot(full_data['Parch'], kde=False, bins = 20)
```

```python
sns.distplot(full_data['Parch'], hist=False, vertical=True)
```

##**JointPlot**

```python
sns.jointplot(data= full_data, x='Pclass', y='Survived', color='olive', kind= 'reg')
```

##**BoxPlot**

```python
sns.boxplot(data= full_data, x='Survived', y='Age')
```

```python
sns.boxplot(data= full_data, x='Survived', hue='Pclass', y='Fare')
```

```python
sns.boxplot(data= full_data, x='Survived', hue='Pclass', y='Fare', palette='Set1')
```

```python
sns.boxplot(data= full_data, x='Survived', hue='Pclass', y='Fare', palette='Set1', linewidth=2)
```

##**Pair Plot**

```python
sns.pairplot(data=full_data, hue='Age', palette='Set1')
```

```python
sns.pairplot(data=full_data, hue='Survived', palette='Set1')
```

```python
sns.pairplot(data=full_data, hue='Sex', palette='Set1')
```

```python
sns.pairplot(data=full_data, hue='Parch', palette='Set1')
```
