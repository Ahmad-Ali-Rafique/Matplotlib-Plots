#**DATA VISULLIZATION USING MATPLOTLIB**

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

##**Line Plot**

```python
plt.plot(x,y)
#Adding Title and Labels
plt.title('Line Plot')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.show()
```

```python
# Changing Line Aesthetics
plt.plot(x,y, color='g', linestyle=':', linewidth='2')
plt.show()
```

```python
# Adding two lines in the same plot
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1, color='b', linestyle='-.', linewidth='2')
plt.plot(x,y2, color='r', linestyle=':', linewidth='3')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# add grid in plot
plt.grid(True)
plt.show()
```

```python
# Adding sub-plots
plt.subplot(1, 2, 1)
plt.plot(x,y1, color='g', linestyle='-.', linewidth='2')
plt.subplot(1, 2, 2)
plt.plot(x,y2, color='r', linestyle=':', linewidth='3')
plt.show()
```

##**Bar Plot**

```python
student={'ahmad':77, 'mujtabah':44, 'zohaib':66}
names=list(student.keys())
numbers=list(student.values())
```

```python
plt.bar(names, numbers)
plt.show()
```

```python
# Adding Title and Labels
plt.bar(names, numbers)
plt.title('Obtained Marks')
plt.xlabel('Names')
plt.ylabel('numbers')
plt.grid(True)
plt.show()
```

```python
# Horizontal Bar Plot
plt.barh(names, numbers, color='g')
plt.title('Obtained Marks')
plt.xlabel('Names')
plt.ylabel('numbers')
plt.grid(True)
plt.show()
```

##**Scatter Plot**

```python
# Creating a basic scatter-plot
x=(10, 20, 30, 40, 50, 60, 70, 80, 90)
y=(4,5,6,7,7,5,4,3,3)
plt.scatter(x, y)
plt.show()
```

```python
# Changing Mark Aesthetics
plt.scatter(x, y, marker="*", c='r', s=50)
plt.show()
```

```python
# Adding two markers in the same plot
x=(10, 20, 30, 40, 50, 60, 70, 80, 90)
y1=(4,5,6,7,7,5,4,3,3)
y2=(3,9,1,6,4,2,5,8,7)
plt.scatter(x,y1,marker=",", c='b', s=100)
plt.scatter(x,y2, marker="*", c='r', s=50)
plt.show()
```

```python
# Adding sub-plots
x=(10, 20, 30, 40, 50, 60, 70, 80, 90)
y1=(4,5,6,7,7,5,4,3,3)
y2=(3,9,1,6,4,2,5,8,7)
plt.subplot(1, 2, 1)
plt.scatter(x,y1,marker=",", c='b', s=100)
plt.subplot(1, 2, 2)
plt.scatter(x,y2, marker="*", c='r', s=50)
plt.show()
```

##**Histogram**

```python
data=(5,6,7,5,3,3,3,6,7,8,6,5,4,3,3,2,4,5,6,7,8,6,5,4,3)
# Making Histogram
plt.hist(data)
plt.show()
```

```python
# Changing Aesthetics
plt.hist(data,  color='r', bins= 4)
plt.show()
```

##**Box-Plot**

```python
# Creating data
one=(1,2,3,4,5,6,7,8,9)
two=(1,2,3,4,5,4,3,2,1)
three=(6,7,8,9,2,3,4,6,7)
data=[one, two, three]
# Making Plot
plt.boxplot(data)
plt.show()
```

##**Violin-Plot**

```python
# Creating data
one=(1,2,3,4,5,6,7,8,9)
two=(1,2,3,4,5,4,3,2,1)
three=(6,7,8,9,2,3,4,6,7)
data=[one, two, three]
# Making Plot
plt.violinplot(data, showmedians=True)
plt.show()
```

##**Pie-Chart**

```python
# Creating data
fruits=('mango', 'apple', 'banana', 'melon')
quantity=(30, 50, 75,73)
# Making Plot
plt.pie(quantity, labels=fruits)
plt.show()
```

```python
# Changing Aesthetics
plt.pie(quantity, labels=fruits, autopct ='%0.1f%%', colors=['green', 'blue', 'orange', 'red'])
plt.show()
```

##**DoughNut-Chart**

```python
# Creating data
fruits=('mango', 'apple', 'banana', 'melon')
quantity=(30, 50, 75,73)
# Making Plot
plt.pie(quantity, labels=fruits,radius=2)
plt.pie([1],colors=['w'], radius=1)
plt.show
```
